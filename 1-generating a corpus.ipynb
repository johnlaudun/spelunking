# LLM Spelunking in Search of Folklore

This project aims to explore the possibilities for using large language models to uncover emergent folklore. 

```python
import requests
import json
from openai import OpenAI
```

```python
keys = json.load(open('../apikeys.json'))
or_key = keys['OpenRouter']["key"]
```

```python
response = requests.post(
  url="https://openrouter.ai/api/v1/chat/completions",
  headers={
    "Authorization": f"Bearer {or_key}"
  },
  data=json.dumps({
    "model": "openai/gpt-5.2", # Optional
    "messages": [
      {
        "role": "user",
        "content": "What is the meaning of life?"
      }
    ]
  })
)
# print(data["messages"]["content"])
print(response)
```

```python
client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=f"{or_key}"
)

completion = client.chat.completions.create(
  extra_headers={
    "HTTP-Referer": "<YOUR_SITE_URL>", # Optional. Site URL for rankings on openrouter.ai.
    "X-Title": "<YOUR_SITE_NAME>", # Optional. Site title for rankings on openrouter.ai.
  },
  model="openai/gpt-5.2",
  messages=[
    {
      "role": "user",
      "content": "What is the meaning of life?"
    }
  ]
)

print(completion.choices[0].message.content)
```
